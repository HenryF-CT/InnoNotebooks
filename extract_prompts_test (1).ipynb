```python
def create_sdg_dataframe(targets_dict):
    """
    Creates a Pandas DataFrame from SDG targets, with Goal Number, Title, and Prompt.

    Args:
        targets_dict (dict): A dictionary with SDG goals and their targets.

    Returns:
        pandas.DataFrame: A DataFrame with Goal Number, Title, and Prompt columns.
    """
    data = []
    for goal, targets in targets_dict.items():
        # Extract goal number and title
        match = re.match(r"Goal (\d+): (.*)", goal)
        if match:
            goal_number = int(match.group(1))
            goal_title = match.group(2)
        else:
            continue #handle cases where the goal number and title are not in the expected format.

        cleaned_targets = []
        for target in targets:
            cleaned_target = re.sub(r"By\s+\d{4},\s*", "", target)
            cleaned_targets.append(cleaned_target)
        combined_prompt = " ".join(cleaned_targets)
        data.append([goal_number, goal_title, combined_prompt])

    df = pd.DataFrame(data, columns=["Goal Number", "Goal Title", "Prompt"])
    df = df.set_index("Goal Number") #set goal number as index
    return df

# Create the DataFrame and print it
sdg_df = create_sdg_dataframe(sdg_targets)
```
